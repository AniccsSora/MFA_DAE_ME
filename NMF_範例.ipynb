```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn.decomposition import NMF
```

```python
X = np.random.rand(8,18)
plt.axis('off')
plt.imshow(X);
```

```python
model = NMF(n_components=2, init='random', random_state=0)
```

```python
W = model.fit_transform(X)
```

```python
H = model.components_
```

```python
W.shape
```

```python
np.dot(W,H).shape
```

```python
plt.axis('off')
plt.imshow(W);
```

```python
plt.axis('off')
plt.imshow(H);
```

```python
data = np.copy(np.round(W, 2))
cols= data.shape[1] #2
rows= data.shape[0] #4
# data = np.arange(rows * cols).reshape((rows, cols))

# Limits for the extent
x_start = 3.0
x_end = 5.0
y_start = 6.0
y_end = 12.0

extent = [x_start, x_end, y_start, y_end]

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 30}
matplotlib.rc('font', **font)

# The normal figure

fig = plt.figure(figsize=(12, 10))  # 4 3
ax = fig.add_subplot(111)
ax.axis('off')
im = ax.imshow(data, extent=extent, origin='lower', interpolation='None', cmap='viridis')

# Add the text
jump_x = (x_end - x_start) / (0.5 * rows)
jump_y = (y_end - y_start) / (9.0 * cols)
x_positions = np.linspace(start=x_start, stop=x_end, num=cols, endpoint=False)
y_positions = np.linspace(start=y_start, stop=y_end, num=rows, endpoint=False)

for y_index, y in enumerate(y_positions):
    for x_index, x in enumerate(x_positions):
        label = data[y_index, x_index]
        text_x = x + jump_x
        text_y = y + jump_y
        ax.text(text_x, text_y, label, color='black', ha='center', va='center')

fig.colorbar(im)
plt.show()
```
