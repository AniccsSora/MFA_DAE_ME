```python
def hamming_defined(n, M):
    return 0.54-0.46*np.cos((2*np.pi*n)/(M-1))
```

```python
n=2048
windows=[]
for i in range(n):
    windows.append(hamming_defined(i, n))
windows = np.array(windows)

fig, [my, official] = plt.subplots(nrows=1, ncols=2)
my.plot(windows)
official.plot(scipy.signal.hamming(n))
plt.show()
```
